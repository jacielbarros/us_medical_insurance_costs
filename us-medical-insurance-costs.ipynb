### The Dataset

The file **`insurance.csv`** contains data about an insured population from the United States and displays the following attributes:

- **Insured Age**: The age of the insured individual.
- **Insured Gender**: The gender of the insured individual (e.g., male or female).

```python
# Convert the data into a list of dictionaries

data_list = []

with open('insurance.csv', 'r') as insurance_csv:
    read_insurance = csv.DictReader(insurance_csv)
    for row in read_insurance:
        # Convert string values to appropriate data types
        row['age'] = int(row['age'])
        row['sex'] = row['sex']
        row['bmi'] = float(row['bmi'])
        row['children'] = int(row['children'])
        row['smoker'] = row['smoker']
        row['region'] = row['region']
        row['charges'] = float(row['charges'])
        data_list.append(row)

print(data_list)
```

## Data Analysis Framework: The `InsuranceAnalysis` Class

To facilitate the analysis of the U.S. medical insurance costs dataset, we've developed a specialized class, `InsuranceAnalysis`. This class serves as a structured framework to organize our data and analytical methods, providing a clear and modular approach to our analysis.

### Class Structure and Functionality

The `InsuranceAnalysis` class is designed to encapsulate the entire dataset, allowing us to interact with and analyze the data through its methods. Here's how it's structured:

- **Initialization**: The class is initialized with a dataset, typically a list of dictionaries, where each dictionary represents an individual insurance record. This setup allows for easy manipulation and analysis of the data.

- **Methods**: We've implemented various methods within the class to perform specific analyses. For example, the `count_females` method calculates the total number of female individuals in the dataset. Additional methods can be added to extend the class's functionality, such as calculating average costs, identifying trends, or segmenting the data based on different criteria.

### Advantages of Using the `InsuranceAnalysis` Class

- **Modularity**: By encapsulating the data and analysis methods within a class, we achieve a modular design that makes the code more organized and maintainable.
- **Reusability**: The class can be easily reused for different sets of data or extended to include more complex analysis methods, enhancing the flexibility and scalability of our analytical approach.
- **Clarity**: Having a dedicated class for our analysis helps in clearly separating the data handling logic from the analysis logic, making the codebase easier to understand and navigate.

In summary, the `InsuranceAnalysis` class is a central part of our data analysis project, providing a robust and flexible foundation for exploring and analyzing insurance cost data.


```python
class InsuranceAnalysis:
    def __init__(self,data_list):
        self.data = data_list
        self.bmi_categories = {
            "underweight": 0,
            "healthy_weight": 18.5,
            "overweight": 25.0,
            "obesity": 30
        }
        
        self.age_groups = {
            "baby": 0,
            "children": 3,
            "adolescent": 13,
            "young_adult": 18,
            "middle_aged": 40,
            "elder": 60
        }
        
        self.assign_bmi_categories()  # Call this method to categorize BMI on initialization
        self.assign_age_groups()
    
    def assign_age_groups(self):
        for record in self.data:
            age = record['age']
            for group, threshold in sorted(self.age_groups.items(), key=lambda x: x[1], reverse=True):
                if age >= threshold:
                    record['age_group'] = group
                    break
    
    def assign_bmi_categories(self):
        for record in self.data:
            bmi_value = record['bmi']
            for category, threshold in sorted(self.bmi_categories.items(), key=lambda x: x[1], reverse=True):
                if bmi_value >= threshold:
                    record['bmi_category'] = category
                    break
                    
    def median_cost(self):
        charges_list = [record['charges'] for record in self.data]
        charges_list.sort()

        n = len(charges_list)
        mid = n // 2  # Integer division to find the middle index

        if n % 2 == 0:  # If even number of elements
            median = (charges_list[mid - 1] + charges_list[mid]) / 2
        else:  # If odd number of elements
            median = charges_list[mid]

        return round(median,2)
    
    def avg_cost_by_region(self):
        cost_by_region ={}
        for record in self.data:
            region = record['region']
            if region not in cost_by_region:
                cost_by_region[region] = []
        for record in self.data:
            region = record['region']
            costs = record['charges']
            cost_by_region[region].append(costs)
            
        avg_cost = {
           region: round(sum(costs) / len(costs),2) if costs else 0
           for region, costs in cost_by_region.items() 
        }
        
        return avg_cost       
    
    
    def avg_cost_by_children(self):
        # Initialize a dictionary with empty lists for each possible number of children
        cost_by_children = {}
        for record in self.data:
            children = record['children']
            if children not in cost_by_children:
                cost_by_children[children] = []

        # Append costs to the corresponding list of children count
        for record in self.data:
            children = record['children']
            cost = record['charges']
            cost_by_children[children].append(cost)

        sorted_cost_by_children = sorted(cost_by_children.items())
        average_costs = {
            children: round(sum(costs) / len(costs), 2) if costs else 0
            for children, costs in sorted_cost_by_children
        }

        return average_costs


    def avg_cost(self):
        total_avg = sum(record['charges'] for record in self.data)
        avg = float(total_avg) / len(self.data)
        return round(avg,2)

    def count_females(self):
        total_females = sum(1 for record in self.data if record['sex'] == 'female')
        return total_females
    def count_males(self):
        total_males = sum(1 for record in self.data if record['sex'] == 'male')
        return total_males
    def total_female_cost(self):
        female_cost = sum(record['charges'] for record in self.data if record['sex'] == 'female')
        return round(female_cost,2)
    def total_male_cost(self):
        male_cost = sum(record['charges'] for record in self.data if record['sex'] == 'male')
        return round(male_cost,2)
    def avg_female_cost(self):
        female_cost = self.total_female_cost()
        female_number = self.count_females()
        female_avg_cost = female_cost / female_number
        return round(female_avg_cost,2)
    def avg_male_cost(self):
        male_cost = self.total_male_cost()
        male_number = self.count_males()
        male_avg_cost = male_cost / male_number
        return round(male_avg_cost,2)
    def smokers_count(self):
        total_smokers = sum(1 for i in self.data if i['smoker'] == 'yes')
        return total_smokers
    def non_smokers_count(self):
        total_non_smokers = sum(1 for i in self.data if i['smoker'] == 'no')
        return total_non_smokers
    def smoker_avg_cost(self):
        total_cost = sum(i['charges'] for i in self.data if i['smoker'] == 'yes')
        smokers_number = self.smokers_count()
        average_cost = total_cost / smokers_number
        return round(average_cost,2)
    
    def non_smoker_avg_cost(self):
        total_cost = sum(i['charges'] for i in self.data if i['smoker'] == 'no')
        non_smokers_number = self.non_smokers_count()
        average_cost = total_cost / non_smokers_number
        return round(average_cost,2)
        
    
    def bmi_underweight_count(self):
        total_underweight = sum(1 for i in self.data if i['bmi_category'] == 'underweight')
        return total_underweight
    def bmi_healthyweight_count(self):
        total_healthyweight = sum(1 for i in self.data if i['bmi_category'] == 'healthy_weight')
        return total_healthyweight
    def bmi_overweight_count(self):
        total_overweight = sum(1 for i in self.data if i['bmi_category'] == 'overweight')
        return total_overweight
    def bmi_obese_count(self):
        total_obese = sum(1 for i in self.data if i['bmi_category'] == 'obesity')
        return total_obese
    def average_cost_by_bmi_category(self):
        category_costs = {category: [] for category in self.bmi_categories.keys()}

        for record in self.data:
            category = record['bmi_category']
            charge = record['charges']
            category_costs[category].append(charge)

        average_costs = {
            category: round(sum(costs) / len(costs), 2) if costs else 0
            for category, costs in category_costs.items()
        }

        return average_costs
    
    def age_baby_count(self):
        total_babies = sum(1 for i in self.data if i['age_group'] == 'baby')
        return total_babies
    def age_children_count(self):
        total_children = sum(1 for i in self.data if i['age_group'] == 'children')
        return total_children
    def age_adolescent_count(self):
        total_adolescent = sum(1 for i in self.data if i['age_group'] == 'adolescent')
        return total_adolescent
    def age_young_adult_count(self):
        total_young_adult = sum(1 for i in self.data if i['age_group'] == 'young_adult')
        return total_young_adult
    def age_middle_aged_count(self):
        total_middle_aged = sum(1 for i in self.data if i['age_group'] == 'middle_aged')
        return total_middle_aged
    def age_elder_count(self):
        total_elder = sum(1 for i in self.data if i['age_group'] == 'elder')
        return total_elder
    def average_cost_age_group(self):
        group_costs = {group: [] for group in self.age_groups.keys()}

        for record in self.data:
            group = record['age_group']
            charge = record['charges']
            group_costs[group].append(charge)

        average_costs = {
            group: round(sum(costs) / len(costs), 2) if costs else 0
            for group, costs in group_costs.items()
        }

        return average_costs

analysis = InsuranceAnalysis(data_list)

print(data_list)
print(analysis.data)
```

## BMI Categories Disclaimer

Body Mass Index (BMI) is a simple calculation using a person's height and weight. While BMI is a useful tool for identifying potential health risks related to weight, it does not directly measure body fat or individual health conditions. Here is a brief overview of the BMI categories used in this analysis:

- **Underweight (BMI less than 18.5)**: This category may indicate that an individual is undernourished or has other health conditions that lead to a low body weight.

- **Healthy Weight (BMI 18.5 to 24.9)**: Generally, individuals in this category have a lower risk of developing weight-related health issues.

- **Overweight (BMI 25 to 29.9)**: This category may indicate a higher risk of developing health conditions, such as heart disease, high blood pressure, or type 2 diabetes, related to excess weight.

- **Obesity (BMI 30 or greater)**: Individuals in this category are at an increased risk for a range of health conditions, including heart disease, high blood pressure, type 2 diabetes, and certain types of cancer.

Please note that BMI is just one of many factors related to health and should not be used as a sole indicator. Consult a healthcare provider for a comprehensive health assessment.
